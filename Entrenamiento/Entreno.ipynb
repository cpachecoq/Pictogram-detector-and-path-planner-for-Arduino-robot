```python
from ultralytics import YOLO

DATA_YAML = "/content/drive/MyDrive/Pictogram_Proyect/1_dataset/pictos7_aug_x6/data.yaml"

model = YOLO("yolov10s.pt")  # o yolov10n.pt si quieres más ligero

results = model.train(
    data=DATA_YAML,
    epochs=150,
    imgsz=768,
    batch=16,        # en A100 puedes subir (48/64) si no revienta
    lr0=0.003,
    freeze=15,       # transfer learning estricto
    patience=75,
    optimizer="AdamW",
    verbose=True,
    project="/content/drive/MyDrive/Pictogram_Proyect/2_training/runs",
    name="pictos7_yolov10s_tl"
)
```
